```python
import random

# Assuming df is your DataFrame
df['VendorID'] = [str(random.randint(10**8, 10**9 - 1)) for _ in range(len(df))]

# If you want to ensure the new IDs are unique within the column
unique_vendor_ids = set(df['VendorID'])
while len(unique_vendor_ids) < len(df):
    unique_vendor_ids.add(str(random.randint(10**8, 10**9 - 1)))

df['VendorID'] = list(unique_vendor_ids)

# If you want the 'VendorID' column to be integers, convert it after generating the IDs
df['VendorID'] = df['VendorID'].astype(int)
```

```python
df.to_csv('original_dataset.csv', index=False)
```

```python
df['tpep_pickup_datetime'] = pd.to_datetime(df.tpep_pickup_datetime)
df['tpep_dropoff_datetime'] = pd.to_datetime(df.tpep_dropoff_datetime)
```

```python
Datetime_df = df[['VendorID','tpep_pickup_datetime','tpep_dropoff_datetime']]
Datetime_df
```

```python
Datetime_df = df[['VendorID','tpep_pickup_datetime','tpep_dropoff_datetime']].reset_index(drop=True)
Datetime_df['tpep_pickup_datetime'] = Datetime_df['tpep_pickup_datetime']
Datetime_df['pickup_hour'] = Datetime_df['tpep_pickup_datetime'].dt.hour
Datetime_df['pickup_day'] = Datetime_df['tpep_pickup_datetime'].dt.day
Datetime_df['pickup_month'] = Datetime_df['tpep_pickup_datetime'].dt.month
Datetime_df['pickup_year'] = Datetime_df['tpep_pickup_datetime'].dt.year
Datetime_df['pickup_weekday'] = Datetime_df['tpep_pickup_datetime'].dt.weekday
```

```python
Datetime_df['tpep_dropoff_datetime'] = Datetime_df['tpep_dropoff_datetime']
Datetime_df['drop_hour'] = Datetime_df['tpep_dropoff_datetime'].dt.hour
Datetime_df['drop_day'] = Datetime_df['tpep_dropoff_datetime'].dt.day
Datetime_df['drop_month'] = Datetime_df['tpep_dropoff_datetime'].dt.month
Datetime_df['drop_year'] = Datetime_df['tpep_dropoff_datetime'].dt.year
Datetime_df['drop_weekday'] = Datetime_df['tpep_dropoff_datetime'].dt.weekday
Datetime_df["datetime_id"] = Datetime_df.index
# Datetime_df = Datetime_df.rename(columns={'tpep_pickupup_datetime': 'datetime_id'}).reset_index(drop=True)
Datetime_df = Datetime_df[['VendorID','datetime_id','tpep_pickup_datetime', 'pickup_hour', 'pickup_day', 'pickup_month', 'pickup_year', 'pickup_weekday',
                             'tpep_dropoff_datetime', 'drop_hour', 'drop_day', 'drop_month', 'drop_year', 'drop_weekday']]
```

```python
passenger_count_dim = df[["VendorID","passenger_count"]].reset_index(drop=True)
passenger_count_dim['passenger_count_id'] = passenger_count_dim.index
passenger_count_dim = passenger_count_dim[['VendorID','passenger_count_id','passenger_count']]

trip_distance_dim = df[['VendorID','trip_distance']].reset_index(drop=True)
trip_distance_dim['trip_distance_id'] = trip_distance_dim.index
trip_distance_dim = trip_distance_dim[['VendorID','trip_distance_id','trip_distance']]
```

```python
rate_code_type = {
    1:"Standard rate",
    2:"JFK",
    3:"Newark",
    4:"Nassau or Westchester",
    5:"Negotiated fare",
    6:"Group ride"
}

rate_code_dim = df[['VendorID','RatecodeID']].reset_index(drop=True)
rate_code_dim['rate_code_id'] = rate_code_dim.index
rate_code_dim['rate_code_name'] = rate_code_dim['RatecodeID'].map(rate_code_type)
rate_code_dim = rate_code_dim[['VendorID','rate_code_id','RatecodeID','rate_code_name']]
```

```python
pickup_df = df[["VendorID","pickup_longitude","pickup_latitude"]].reset_index(drop = True)
pickup_df["pickup_id"] = pickup_df.index
pickup_df = pickup_df[["VendorID","pickup_id","pickup_longitude","pickup_latitude"]]
```

```python
dropoff_df = df[["VendorID","dropoff_longitude","dropoff_latitude"]].reset_index(drop = True)
dropoff_df["dropoff_id"] = dropoff_df.index
dropoff_df = dropoff_df[["VendorID","dropoff_id","dropoff_longitude","dropoff_latitude"]]
```

```python
payment_type_name = {
    1:"Credit card",
    2:"Cash",
    3:"No charge",
    4:"Dispute",
    5:"Unknown",
    6:"Voided trip"
}
payment_df = df[["VendorID","payment_type"]].reset_index(drop=True)
payment_df["Payment_id"] = payment_df.index
payment_df["payment_names"] = payment_df["payment_type"].map(payment_type_name)
payment_df = payment_df[["VendorID","Payment_id","payment_type","payment_names"]]
```

```python
fact_table = df.merge(passenger_count_dim, on='VendorID') \
             .merge(trip_distance_dim, on='VendorID') \
             .merge(rate_code_dim, on='VendorID') \
             .merge(pickup_df, on=['VendorID']) \
             .merge(dropoff_df, on=['VendorID'])\
             .merge(Datetime_df, on=['VendorID']) \
             .merge(payment_df, on='VendorID') \
             [['VendorID', 'datetime_id', 'passenger_count_id',
               'trip_distance_id', 'rate_code_id', 'store_and_fwd_flag', 'pickup_id', 'dropoff_id',
               'Payment_id', 'fare_amount', 'extra', 'mta_tax', 'tip_amount', 'tolls_amount',
               'improvement_surcharge', 'total_amount']]
```
